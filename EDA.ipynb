```python
final_feats = [
    'NU_NOTA_CN', 'NU_NOTA_CH', 'NU_NOTA_LC', 'NU_NOTA_REDACAO',
    'NU_IDADE', 'TP_DEPENDENCIA_ADM_ESC', 'TP_LINGUA',
    'TP_ESCOLA', 'TP_ENSINO', 'TP_ST_CONCLUSAO',
    'IN_TREINEIRO', 'TP_STATUS_REDACAO'
]
```

```python
import matplotlib.pyplot as plt
# Verificar características que tem dados nulos

feats_null = [f for f in final_feats if df_train[f].isnull().any()]

fig, axs = plt.subplots(len(feats_null), 1, figsize=(12,20))

for i,f in enumerate(feats_null):
    ax = axs[i]
    ax.set_title(f)
    df_train[f].hist(ax=ax)

plt.tight_layout()
```
